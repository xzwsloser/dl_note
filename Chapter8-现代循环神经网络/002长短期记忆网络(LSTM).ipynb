# 长短期记录网络(LSTM)
隐变量模型存在长期信息保存(梯度消失)和短期输入缺失(把以往的输入压缩为单个变量 $\mathbf{H}_t$)的问题
## 门控记忆元
`LSTM`网络中引入了记忆元, 可以看成隐状态的一种特殊类型, 和隐状态具有相同的形状, 并且涉及的目的就是用于记录附加的信息 ; 同时为了控制记忆元, 需要一些门, 其中一个门用于从单元中输出条目, 称为输出门 ; 另外一个门用于决定何时将数据读如单元, 称为输入门, 同时需要一种机制重置单元信息, 称为一望无际门, 这一种设计类似于门控循环单元

### 输入门、遗忘门和输出门
类似于门控循环单元, 当前时间步的输入 $\mathbf{X}_t$ 和前一个时间步的隐状态作为数据被送入长短期记录网络的门中, 同时它们由三个具有 `sigmoid` 激活函数的全连接层处理, 以计算输入门、遗忘门和输出门的值, 并且三个门的值都在 $(0,1)$的范围内:

考虑 `LSTM` 循环单元的数学表达, 假设存在$h$个隐藏单元, 批量大小为 $n$, 输入数为 $d$, 因此, 输入为 $\mathbf{X}_t \in \mathbb{R}^{n \times d}$, 前一个时间步的隐状态为 $\mathbf{H}_{t-1} \in \mathbb{n \times h}$, 响应的, 时间步 $t$ 的门定义如下, 输入门为 $\mathbf{I}_t \in \mathbb{R}^{n \times h}$, 遗忘门为 $\mathbf{F}_t \in \mathbb{R}^{n \times h}$, 输出门 $\mathbf{O}_t \in \mathbb{n \times h}$, 计算方法如下:
$$
\begin{aligned}
\mathbf{I}_t &= \sigma(\mathbf{X}_t \mathbf{W}_{xi} + \mathbf{H}_{t-1} \mathbf{W}_{hi} + \mathbf{b}_i) \\
\mathbf{F}_t &= \sigma(\mathbf{X}_t \mathbf{W}_{xf} + \mathbf{H}_{t-1} \mathbf{W}_{hf} + \mathbf{b}_f) \\
\mathbf{O}_t &= \sigma(\mathbf{X}_t \mathbf{W}_{xo} + \mathbf{H}_{t-1} \mathbf{W}_{ho} + \mathbf{b}_o) 
\end{aligned}
$$
其中 $\mathbf{W}_{xi}, \mathbf{W}_{xf}, \mathbf{W}_{xo} \in \mathbb{R}^{d \times h}$ 和 $\mathbf{W}_{hi}, \mathbf{W}_{hf}, \mathbf{W}_{ho} \in \mathbb{R}^{h \times h}$为权重参数, $\mathbf{b}_i, \mathbf{b}_f, \mathbf{b}_o \in \mathbb{R}^{1 \times h}$为偏置参数

### 候选记忆元
候选记忆元的计算方式类似于三个门的计算, 但是使用 `tanh` 作为激活函数, 候选记忆元为 $\tilde{\mathbf{C}}_t \in \mathbb{R}^{n \times h}$, 计算方法如下:
$$
\tilde{\mathbf{C}}_t = \tanh(\mathbf{X}_t \mathbf{W}_{xc} + \mathbf{H}_{t-1} \mathbf{W}_{hc} + \mathbf{b}_c)
$$
其中 $\mathbf{W}_{xc} \in \mathbb{R}^{d \times h}$ 以及 $\mathbf{W}_{hc} \in \mathbb{R}^{h \times h}$为权重参数, $\mathbf{b}_c \in \mathbb{R}^{1 \times h}$ 为偏置参数

### 记忆元
在门控循环单元中, 利用一种机制来控制输入和遗忘(或跳过), 在 `LSTM` 中, 也有两个门用于这样的目的, 输入门 $\mathbf{I}_t$ 控制采用多少来自 $\tilde{\mathbf{C}}_t$的新数据, 同时遗忘门 $\mathbf{F}_t$控制保留多少过去的记忆元 $\mathbf{C}_{t-1} \in \mathbb{R}^{n \times h}$, 所以此时可以得到记忆元的计算方式为:
$$
\mathbf{C}_t = \mathbf{F}_t \odot \mathbf{C}_{t-1} + \mathbf{I}_t \odot \tilde{\mathbf{C}}_t
$$

### 隐状态
考虑隐状态 $\mathbf{H}_t \in \mathbb{R}^{n \times h}$ 的计算方式, 利用输出门限制了 $\mathbf{H}_t$ 的值保持在 $(-1 , 1)$, 计算方式如下:
$$
\mathbf{H}_t = \mathbf{O}_t \odot \tanh(\mathbf{C}_t)
$$
可以发现输出门如果接近 $0$, 只保留记忆元内的所有信息, 输出门接近 $1$, 可以有效把所有记忆信息传递给预测部分

## 总结一下
总结几个变量的作用:
- $\mathbf{F}_t$: 遗忘门, 作为记忆单元 $\mathbf{C}_{t-1}$ 的系数, 反应的是保留多少之前的记忆
- $\mathbf{I}_t$: 输入门, 作为候选单元 $\tilde{\mathbf{C}}_t$ 的系数, 表示当前输入以及之前隐状态的影响
- $\tilde{\mathbf{C}}_t$: 候选状态, 表示当前输入和之前状态的影响
- $\mathbf{O}_t$: 输出门控制最后的输出
- $\mathbf{C}_t$: 当前记忆单元, 用于存储之前记忆

## LSTM实现(造轮子)

```python
# 初始化数据集
batch_size, num_steps = 32, 35
train_iter, vocab = mdl.load_data_time_machine(batch_size, num_steps)
```

### 初始化模型参数

```python
def get_lstm_params(vocab_size, num_hiddens, device):
    num_inputs = num_outputs = vocab_size

    def normal(shape):
        return torch.randn(size=shape, device=device)*0.01

    def three():
        return (normal((num_inputs, num_hiddens)),
                normal((num_hiddens, num_hiddens)),
                torch.zeros(num_hiddens, device=device))

    W_xi, W_hi, b_i = three()  # 输入门参数
    W_xf, W_hf, b_f = three()  # 遗忘门参数
    W_xo, W_ho, b_o = three()  # 输出门参数
    W_xc, W_hc, b_c = three()  # 候选记忆元参数
    # 输出层参数
    W_hq = normal((num_hiddens, num_outputs))
    b_q = torch.zeros(num_outputs, device=device)
    # 附加梯度
    params = [W_xi, W_hi, b_i, W_xf, W_hf, b_f, W_xo, W_ho, b_o, W_xc, W_hc,
              b_c, W_hq, b_q]
    for param in params:
        param.requires_grad_(True)
    return params
```

### 定义模型

```python
# 初始化 LSTM 循环单元状态 (C_t, H_t)
def init_lstm_state(batch_size, num_hiddens, device):
    return (torch.zeros((batch_size, num_hiddens), device=device),
            torch.zeros((batch_size, num_hiddens), device=device))
```

```python
# 定义 LSTM 模型计算方法
def lstm(inputs, state, params):
    [W_xi, W_hi, b_i, W_xf, W_hf, b_f, W_xo, W_ho, b_o, W_xc, W_hc, b_c,
     W_hq, b_q] = params
    (H, C) = state
    outputs = []
    for X in inputs:
        I = torch.sigmoid((X @ W_xi) + (H @ W_hi) + b_i)
        F = torch.sigmoid((X @ W_xf) + (H @ W_hf) + b_f)
        O = torch.sigmoid((X @ W_xo) + (H @ W_ho) + b_o)
        C_tilda = torch.tanh((X @ W_xc) + (H @ W_hc) + b_c)
        C = F * C + I * C_tilda
        H = O * torch.tanh(C)
        Y = (H @ W_hq) + b_q
        outputs.append(Y)
    return torch.cat(outputs, dim=0), (H, C)
```

### 模型训练和预测

```python
vocab_size, num_hiddens = len(vocab), 256
num_epochs, lr = 500, 1
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# 定义模型
model = mdl.RNNModelScratch(vocab_size, num_hiddens, device, get_lstm_params,
                           init_lstm_state, lstm)
mdl.train_ch8(model, train_iter, vocab, lr, num_epochs, device)
```

## 简结实现

```python
num_inputs = vocab_size
lstm_layer = nn.LSTM(num_inputs, num_hiddens)
model = mdl.RNNModel(lstm_layer, len(vocab))
model = model.to(device)
mdl.train_ch8(model, train_iter, vocab, lr, num_epochs, device)
```
