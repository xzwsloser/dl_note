```python
import torch
import math
from torch import nn
import torch.nn.functional as F
from d2l import torch as d2l
```

# 注意力评分函数
对于上一节提到的核回归公式:
$$
\begin{align*}
f(x) &= \sum_{i=1}^{n} \alpha(x, x_i) y_i \\
&= \sum_{i=1}^{n} \frac{\exp\left(-\frac{1}{2}(x - x_i)^2\right)}{\sum_{j=1}^{n} \exp\left(-\frac{1}{2}(x - x_j)^2\right)} y_i \\
&= \sum_{i=1}^{n} \text{softmax}\left(-\frac{1}{2}(x - x_i)^2\right) y_i
\end{align*}
$$
其中 $softmax$ 中的部分就是注意力评分函数, 结合注意力评分函数之后, 整个输出的计算过程如下:

使用数学语言描述, 假设一个查询 $\mathbf{q} \in \mathbb{R}^q$ 和 $m$ 个 `K-V`键值对 $(\mathbf{k}_1, \mathbf{v}_1), \ldots, (\mathbf{k}_m, \mathbf{v}_m)$, 其中 $\mathbf{k}_i \in \mathbb{R}^k, \mathbf{v}_i \in \mathbb{R}^v$, 注意力汇聚函数 $f$ 就被栈时为值的加权和(权重为注意力权重): 
$$
f(\mathbf{q}, (\mathbf{k}_1, \mathbf{v}_1), \ldots, (\mathbf{k}_m, \mathbf{v}_m)) = \sum_{i=1}^{m} \alpha(\mathbf{q}, \mathbf{k}_i) \mathbf{v}_i \in \mathbb{R}^v
$$
注意上式中 $q, k, v$ 都是表示 `Q, K, V` 对应的特征数量, 整个过程只考虑一个样本, 其中查询$\mathbf{q}$和键$\mathbf{k}_i$ 的注意力权重是通过注意力评分函数$\alpha$将两个向量映射为标量(注意力权重), 之后经过 $softmax$ 运算得到:
$$
\alpha(\mathbf{q}, \mathbf{k}_i) = \text{softmax}(a(\mathbf{q}, \mathbf{k}_i)) = \frac{\exp(a(\mathbf{q}, \mathbf{k}_i))}{\sum_{j=1}^m \exp(a(\mathbf{q}, \mathbf{k}_j))} \in \mathbb{R}.
$$
这里选择不同的注意力评分函数可能导致不同的注意力汇聚操作, 下面介绍两种比较流行的注意力评分函数, 并且之后使用这两个函数来实现更加复杂的注意力机制

## 掩蔽softmax操作
和前面的 `Loss` 函数中遮蔽无用的词元的操作一样, 这里一些无关的词元也需要屏蔽, 仅仅对于由意义的词元使用值来获取注意力汇聚, 下面指定一个有效序列长度(词元个数), 以便在计算 `softmax` 时过滤掉超出指定范围的位置, 下面指向掩蔽 `softmax` 操作, 任何超出有效长度的位置都将掩蔽并且设置为 $0$

主要的思路是使用一个很小的负数代替被遮蔽的位置, 这样取指操作之后得到的数字接近 $0$

```python
#@save
def masked_softmax(X, valid_lens):
    """通过在最后一个轴上掩蔽元素来执行softmax操作"""
    # X:3D张量，valid_lens:1D或2D张量
    if valid_lens is None:
        return F.softmax(X, dim=-1)
    else:
        shape = X.shape
        if valid_lens.dim() == 1:
            valid_lens = torch.repeat_interleave(valid_lens, shape[1])
        else:
            valid_lens = valid_lens.reshape(-1)
        # 最后一轴上被掩蔽的元素使用一个非常大的负值替换，从而其softmax输出为0
        X = d2l.sequence_mask(X.reshape(-1, shape[-1]), valid_lens,
                              value=-1e6)
        return F.softmax(X.reshape(shape), dim=-1)
```

```python
# 考虑 2*4 矩阵表示样本, 分别进行遮蔽
masked_softmax(torch.rand(2, 2, 4), torch.tensor([2, 3]))
```

```python
# 使用二维张量, 为矩阵样本中的每一行指定有效长度
masked_softmax(torch.rand(2, 2, 4), torch.tensor([[1, 2],[3, 4]]))
```

## 加性注意力
当查询和键是不同长度的矢量的时候, 可以使用加性注意力作为评分函数, 给定查询 $\mathbf{q} \in \mathbb{R}^q$ 以及键 $\mathbf{k} \in \mathbb{R}^k$, 加性注意力评分函数为:
$$
a(\mathbf{q}, \mathbf{k}) = \mathbf{w}_v^T \tanh(\mathbf{W}_q\mathbf{q} + \mathbf{W}_k\mathbf{k}) \in \mathbb{R}
$$
其中可学习的参数 $\mathbf{W}_q \in \mathbb{R}^{h \times q}$, $\mathbf{W}_k \in \mathbb{R}^{h \times k}$ 以及 $\mathbf{w}_v \in \mathbb{R}^h$

```python
#@save
class AdditiveAttention(nn.Module):
    """加性注意力"""
    def __init__(self, key_size, query_size, num_hiddens, dropout, **kwargs):
        super(AdditiveAttention, self).__init__(**kwargs)
        self.W_k = nn.Linear(key_size, num_hiddens, bias=False)
        self.W_q = nn.Linear(query_size, num_hiddens, bias=False)
        self.w_v = nn.Linear(num_hiddens, 1, bias=False)
        self.dropout = nn.Dropout(dropout)

    def forward(self, queries, keys, values, valid_lens):
        queries, keys = self.W_q(queries), self.W_k(keys)
        # 在维度扩展后，
        # queries的形状：(batch_size，查询的个数，1，num_hidden)
        # key的形状：(batch_size，1，“键－值”对的个数，num_hiddens)
        # 使用广播方式进行求和
        features = queries.unsqueeze(2) + keys.unsqueeze(1)
        features = torch.tanh(features)
        # self.w_v仅有一个输出，因此从形状中移除最后那个维度。
        # scores的形状：(batch_size，查询的个数，“键-值”对的个数)
        scores = self.w_v(features).squeeze(-1)
        self.attention_weights = masked_softmax(scores, valid_lens)
        # values的形状：(batch_size，“键－值”对的个数，值的维度)
        return torch.bmm(self.dropout(self.attention_weights), values)
```

下面测试 `AdditiveAttention` 类, 其中查询、键和值的形状为 `(批量大小, 步数或者词元序列长度, 特征大小)`, 实际输出大小为 `(2, 1, 20)、(2, 10, 2)、(2, 10, 4)`, 注意力汇聚输出形状为 `(批量大小, 查询的步数, 值维度)`

```python
queries, keys = torch.normal(0, 1, (2, 1, 20)), torch.ones((2, 10, 2))
# values 的小批量, 两个值矩阵相同
values = torch.arange(40, dtype=torch.float32).reshape(1, 10, 4).repeat(2, 1, 1)
valid_lens = torch.tensor([2, 6])
attention = AdditiveAttention(key_size=2, query_size=20, num_hiddens=8,
                             dropout=0.1)
attention.eval()
output = attention(queries, keys, values, valid_lens)
output, output.shape
```

```python
# 展示加性注意力下的热力图
d2l.show_heatmaps(attention.attention_weights.reshape((1, 1, 2, 10)), 
                  xlabel='Keys', ylabel='Queries')
```

## 缩放点积注意力
点积操作需要查询和键具有相同的长度$d$(其实就是两者的特征大小相等), 假设查询和键的所有元素都是独立的随机变量, 并且都满足零均值和单位方差, 那么两个向量的点积的均值为 $0$, 并且方差为 $d$, 无论保证向量长度如何, 点积的方差在不考虑向量长度的情况下仍然是 $1$, 所以可以把点积除 $\sqrt d$, 所以缩放点积注意力评分函数为:
$$
a(\mathbf{q}, \mathbf{k}) = \frac {\mathbf{q}^T \mathbf{k}}{\sqrt d}
$$
可以从小批量的角度来考虑提高效率, 比如基于 $n$ 个查询和 $m$ 个键值对来计算注意力, 其中查询和键的长度为 $d$, 值的长度为$v$, 查询$\mathbf{Q} \in \mathbb{R}^{n \times d}$, 键 $\mathbf{K} \in \mathbb{R}^{m \times d}$ 和值 $\mathbf{V} \in \mathbb{R}^{m \times v}$ 的缩放点积注意力是:
$$
softmax(\frac{\mathbf{Q}\mathbf{K}^T}{\sqrt d})\mathbf{V} \in \mathbb{R}^{n \times v}
$$

```python
#@save
class DotProductAttention(nn.Module):
    """缩放点积注意力"""
    def __init__(self, dropout, **kwargs):
        super(DotProductAttention, self).__init__(**kwargs)
        self.dropout = nn.Dropout(dropout)
    # queries 形状 -> (batch_size, 查询的个数, d)
    # keys    形状 -> (batch_size, K-V对的个数, d)
    # values  形状 -> (batch_size, K-V对的个数, 值的维度)
    # valid_len 形状 -> (batch_size,) 或者 (batch_size, 查询个数)
    def forward(self, queries, keys, values, valid_lens=None):
        d = queries.shape[-1]
        # 设置 transpose_b=True交换keys的最后两个维度
        scores = torch.bmm(queries, keys.transpose(1, 2)) / math.sqrt(d)
        self.attention_weights = masked_softmax(scores, valid_lens)
        return torch.bmm(self.dropout(self.attention_weights), values)
```

```python
# 测试 缩放点积注意力
queries = torch.normal(0, 1, (2, 1, 2)) # 注意特侦数量
attention = DotProductAttention(dropout=0.5)
attention.eval()
attention(queries, keys, values, valid_lens)
```

```python
# 缩放点积注意力
d2l.show_heatmaps(attention.attention_weights.reshape((1, 1, 2, 10)),
                  xlabel='Keys', ylabel='Queries')
```
