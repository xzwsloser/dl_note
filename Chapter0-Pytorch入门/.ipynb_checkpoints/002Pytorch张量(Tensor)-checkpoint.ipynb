```python
# 2.  torch.stack, 
tensor_2d = torch.tensor([
    [-9, 4, 2, 5, 7],
    [3, 0, 12, 8, 6],
    [1, 23, -6, 45, 2],
    [22, 3, -1, 72, 6]
])

print("2D Tensor (Matrix):\n", tensor_2d)
print("Shape:", tensor_2d.shape)  # 

#  3D 
tensor_3d = torch.stack([tensor_2d, tensor_2d + 10, tensor_2d - 5])  #  3  2D 
print("3D Tensor (Cube):\n", tensor_3d)
print("Shape:", tensor_3d.shape)  # 
```

```python
# 1.  2D 
tensor = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
# 2. 
print('Tensor: \n', tensor)
print('Shape:', tensor.shape)
print('Size:', tensor.size())
print('Data Type:', tensor.dtype)
print('Device:', 'cuda' if tensor.is_cuda else 'cpu')
print('Dim:', tensor.dim())
print('Total Elements:',  tensor.numel())
print('T: \n', tensor.T)
print('Requre grad:', tensor.requires_grad)
print('Is Contiguous:', tensor.is_contiguous)
```

##  Numpy 
- `torch.from_numpy(ndarray)`:  `numpy` 
- `x.numpy`:  `numpy` ( `cpu` )
 `ndarray` ,  `clone` 

```python
# 1. NumPy  PyTorch 
print("1. NumPy  PyTorch ")
numpy_array = np.array([[1, 2, 3], [4, 5, 6]])
print("NumPy :\n", numpy_array)

#  torch.from_numpy()  NumPy 
tensor_from_numpy = torch.from_numpy(numpy_array)
print(" PyTorch :\n", tensor_from_numpy)

#  NumPy 
numpy_array[0, 0] = 100
print(" NumPy :\n", numpy_array)
print("PyTorch :\n", tensor_from_numpy)

# 2. PyTorch  NumPy 
print("\n2. PyTorch  NumPy ")
tensor = torch.tensor([[7, 8, 9], [10, 11, 12]], dtype=torch.float32)
print("PyTorch :\n", tensor)

#  tensor.numpy()  NumPy 
numpy_from_tensor = tensor.numpy()
print(" NumPy :\n", numpy_from_tensor)

#  NumPy 
tensor[0, 0] = 77
print(" PyTorch :\n", tensor)
print("NumPy :\n", numpy_from_tensor)

# 3. 
print("\n3.  clone() ")
tensor_independent = torch.tensor([[13, 14, 15], [16, 17, 18]], dtype=torch.float32)
numpy_independent = tensor_independent.clone().numpy()  #  clone 
print(":\n", tensor_independent)
tensor_independent[0, 0] = 0  # 
print(":\n", tensor_independent)
print("NumPy :\n", numpy_independent)
```
