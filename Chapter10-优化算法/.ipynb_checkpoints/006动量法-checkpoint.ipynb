# 动量法
## 基础
### 泄漏平均值
小批量随机梯度下降可以使用如下公式计算:
$$
\mathbf{g}_{t, t-1} = \partial_w \frac {1}{|\mathcal{B}_t|} \sum_{i \in \mathcal{B}_t} f(\mathbf{x}_i, \mathbf{w}_{t-1}) = \frac {1}{|\mathcal{B}_t|} \sum_{i \in \mathcal{B}_t} \mathbf{h}_{i, t-1}
$$
这里使用 $\mathbf{h}_{i, t-1} = \partial_w f(\mathbf{x}_i, \mathbf{w}_{t-1})$ 作为样本 $i$ 的随机梯度下降, 使用时间步$t-1$ 是更新的权重$t - 1$, 下面考虑泄漏平均值:
$$
\mathbf{v}_t = \beta \mathbf{v}_{t-1} + \mathbf{g}_{t, t-1}
$$
其中 $\beta \in (0,1)$, 由此可以把瞬时梯度替换为过去梯度的平均值, $\mathbf{v}$ 称为动量, 这一个变量累加了过去的梯度, 比如展开 $\mathbf{v}_t$ 可以得到:
$$
\mathbf{v}_t = \beta^2 \mathbf{v}_{t-2} + \beta \mathbf{g}_{t-1, t-2} + \mathbf{g}_{t, t-1} = \ldots = \sum_{\tau = 0}^{t - 1} \beta^{\tau} \mathbf{g}_{t - \tau, t - \tau - 1}
$$
其中如果 $\beta$ 较大表示长期平均值(也就是过去的梯度的影响比较大), 同时如果 $\beta$ 较小, 那么仅仅是对于梯度法略有修正, 新的梯度替换不再指向所特定实例下降最都的方向, 而是指向过去梯度的加权平均值的方向

**动量法核心:** 使用以往梯度的加权平均来表示目前的梯度 

### 梯度下降条件不佳的情况
考虑函数 $f(x_1, x_2) = 0.1x_1^2 + 2x_2^2$, 该函数在 $f(0, 0)$ 的位置取得最小值, 并且在 $x_1$ 的方向上比较平坦, 新函数在执行梯度下降的过程中可能存在一些问题: 
- lr 设置比较小的情况下, 可以保证在 $x_2$ 方向上不发散, 但是在 $x_1$ 的方向上收敛慢
- lr 设置比较大的情况下, 可能会发生在 $x_2$ 方向上的振荡

```python
eta = 0.4
def f_2d(x1, x2):
    return 0.1 * x1 ** 2 + 2 * x2 ** 2
def gd_2d(x1, x2, s1, s2):
    return (x1 - eta * 0.2 * x1, x2 - eta * 4 * x2, 0, 0)
d2l.show_trace_2d(f_2d, d2l.train_2d(gd_2d))
```

```python
# 设置学习率比较大的情况
eta = .6
d2l.show_trace_2d(f_2d, d2l.train_2d(gd_2d))
```

### 动量法
在上述例子中利用动量法, 可能在计算过去的平均梯度的效果比较好, 在 $x_1$ 方向上, 会聚合比较对齐的梯度, 从而增加每一步中覆盖的距离, 同时在梯度振荡的$x_2$方向上, 由于相互啼笑了对方的振荡, 聚合梯度将减小步长大小, 使用 $\mathbf{v}_t$ 而不是梯度 $\mathbf{g}_t$ 生成如下更新式:
$$
\begin{align*}
\boldsymbol{v}_t \gets \beta \boldsymbol{v}_{t-1} + \boldsymbol{g}_{t,t-1}, \\
\boldsymbol{x}_t \gets \boldsymbol{x}_{t-1} - \eta_t \boldsymbol{v}_t.
\end{align*}
$$
如果其中 $\beta = 0$, 表示不考虑过去梯度加权和的影响, 等价于常规的梯度下降

```python
# 动量法的实现
def momentum_2d(x1, x2, v1, v2):
    v1 = beta * v1 + 0.2 * x1
    v2 = beta * v2 + 4 * x2
    return x1 - eta * v1, x2 - eta * v2, v1, v2

eta, beta = 0.6, 0.5
d2l.show_trace_2d(f_2d, d2l.train_2d(momentum_2d))
```

```python
eta, beta = 0.6, 0.25
d2l.show_trace_2d(f_2d, d2l.train_2d(momentum_2d))
```

### 有效样本权重
由于 $\mathbf{v}_t = \sum_{\tau = 0}^{t - 1} \beta^{\tau} \mathbf{g}_{t-\tau, t-\tau-1}$, 并且极限情况下 $\sum_{\tau = 0}^{\infty} \beta^{\tau} = \frac {1}{1 - \beta}$, 所以选取不同的 $\beta$ 可能产生不同影响

```python
d2l.set_figsize()
betas = [0.95, 0.9, 0.6, 0]
for beta in betas:
    x = torch.arange(40).detach().numpy()
    d2l.plt.plot(x, beta ** x, label=f'beta = {beta:.2f}')
d2l.plt.xlabel('time')
d2l.plt.legend();
```

### 实现
深度学习框架中的优化求解器已经构建了动量法, 设置匹配参数就可以产生比较类似的轨迹, 这里可以参考官方文档中对于 `torch.optim.SGD` 的描述:


```python
data_iter, _ = d2l.get_data_ch11(batch_size=10)
trainer = torch.optim.SGD
d2l.train_concise_ch11(trainer, {'lr': 0.005, 'momentum': 0.9}, data_iter)
```

## 总结
- 动量法使用过去梯度的平均值来替换梯度, 从而加快收敛速度
- 对于无噪声梯度下降和嘈杂随机梯度下降, 动量法适用
- 动量法可以防止在随机梯度下降的优化过程中停滞的问题(过去的梯度不为 $0$)
- 由于对过去的数据进行指数降权, 有小梯度数为 $\frac {1}{1 - \beta}$
- 在凸二次问题中, 可以对东梁发进行明确详细的分析
- 动量法的实现比较简单, 但需要我们存储额外的状态向量(动量 $\mathbf{v}$)
