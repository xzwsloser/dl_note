# RMSProp算法
`Adagrad`对于深度学习中的非凸问题可能效果不太好, `RMSProp`算法类似于动量法 + `AdaGrad`算法, 利用泄漏平均值来更新 $\mathbf{s}_t$

## 算法
更新过程如下:
$$
\begin{align*}
\mathbf{s}_t &\leftarrow \gamma \mathbf{s}_{t-1} + (1 - \gamma) \mathbf{g}_t^2, \\
\mathbf{x}_t &\leftarrow \mathbf{x}_{t-1} - \frac{\eta}{\sqrt{\mathbf{s}_t + \epsilon}} \odot \mathbf{g}_t.
\end{align*}
$$
常数 $\epsilon > 0$ 通常可以设置为 $10^{-6}$, 从而确保除以 $0$ 或者补步长过大而造成一定的影响, 或者可以自由控制学习率 $\eta$, 展开 $\mathbf{s}_t$ 可以得到:
$$
\mathbf{s}_t = (1 - \gamma) \mathbf{g}_t^2 + \gamma \mathbf{s}_{t-1} = (1 - \gamma)(\mathbf{g}_t^2 + \gamma \mathbf{g}_{t-1}^2 + \gamma^2 \mathbf{g}_{t-2} + \ldots)
$$
下面可以观测随着 $\gamma$ 的不同, 各阶导数所占的权重:

```python
d2l.set_figsize()
gammas = [0.95, 0.9, 0.8, 0.7]
for gamma in gammas:
    x = torch.arange(40).detach().numpy()
    d2l.plt.plot(x, (1-gamma) * gamma ** x, label=f'gamma = {gamma:.2f}')
d2l.plt.xlabel('time');
```

## 实现
这里可以使用 `PyTorch API` 中提供的 `torch.optim.RMSRrop` 算法来训练模型


```python
data_iter, _ = d2l.get_data_ch11(batch_size=10)
trainer = torch.optim.RMSprop
d2l.train_concise_ch11(trainer, {'lr': 0.01, 'alpha': 0.9}, data_iter)
```
