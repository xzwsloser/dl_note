# AdaGrad算法
## 稀疏特征和学习率
稀疏特征: 偶尔出现的特征, 这一个问题在 `NLP` 任务中比较常见, 并且当这些不常见的特征出现的时候, 与其相关的参数才会得到有意义的更新, 但是如果设置一般的学习率(比如与 $t$ 相关或者常数), 学习率要么对于常见特征而言降低太慢, 要么对于不常见特征而言降低太快

解决这一个问题的方法就是记录看到特定特征的次数, 并且用于调整学习率, 比如使用大小为 $\eta_i = \frac {\eta_0}{\sqrt{s(i,t) + c}}$, 其中 $s(i, t)$记录了截至$t$时观察到功能$i$的次数, 其中 $\eta_i$ 针对于特征$i$

`AdaGrad`算法通过计数器$s(i,t)$替换先前观察得到的梯度的平方和来解决这一个问题, 并且使用$s(i,t+1) = s(i,t) + (\partial f(\mathbf{x}))^2$来调整学习率, 优点:
1. 不需要决定梯度何时足够大
2. 随着梯度大小子读懂变换

## 算法
首先, 使用$\mathbf{s}_t$来累加过去的梯度方差, 公式如下:
$$
\begin{align*}
\mathbf{g}_t &= \partial_{\mathbf{w}} l(y_t, f(\mathbf{x}_t, \mathbf{w})), \\
\mathbf{s}_t &= \mathbf{s}_{t - 1} + \mathbf{g}_t^2, \\
\mathbf{w}_t &= \mathbf{w}_{t - 1} - \frac{\eta}{\sqrt{\mathbf{s}_t + \epsilon}} \cdot \mathbf{g}_t.
\end{align*}
$$
其实有一点类似于动量法, 同样是对于之前的梯度进行累加, 这样可以保证梯度下降的轨迹更加平滑

```python
# 测试 AdaGrad 算法, f(x) = 0.1(x_1)^2 + 2(x_2)^2
def adagrad_2d(x1, x2, s1, s2):
    eps = 1e-6
    g1, g2 = 0.2 * x1, 4 * x2
    s1 += g1 ** 2
    s2 += g2 ** 2
    x1 -= eta / math.sqrt(s1 + eps) * g1
    x2 -= eta / math.sqrt(s2 + eps) * g2
    return x1, x2, s1, s2
def f_2d(x1, x2):
    return 0.1 * x1 ** 2 + 2 * x2 ** 2
eta = 0.4
d2l.show_trace_2d(f_2d, d2l.train_2d(adagrad_2d))
```

```python
# 提高学习率, 可以发现学习率的降低比较剧烈
eta = 2
d2l.show_trace_2d(f_2d, d2l.train_2d(adagrad_2d))
```

## 实现
使用 `PyTorch API` 中提供的 `AdaGrad` 算法来训练模型


```python
data_iter, _ = d2l.get_data_ch11(batch_size=10)
trainer = torch.optim.Adagrad
d2l.train_concise_ch11(trainer, {'lr': 0.1}, data_iter)
```
