## 独热编码(One-Hot)
在 `train_iter`中, 每一个词元表示为一个数字索引, 将这些索引直接输入神经网络中可能到时学习比较困难, 通常可以把每一个词元标识为更具表现力的特征向量, 最简单的表示就是独热编码

也就是把每一个索引映射为相互不同的单位向量, 假设词表中不同词元的数量为 $N$(`len(vocab)`), 词元索引的范围为 $0$ 到 $N - 1$, 假设词元的索引是整数 $i$, 那么我们将创建一个长度为 $N$ 的全 $0$ 向量, 并且将第 $i$ 处的元素设置为 $1$, `PyTorch` 中有相应的 `API` 构造 `One-Hot` 变量

```python
# 独热编码
F.one_hot(torch.tensor([0, 2]), len(vocab))
```

```python
# 注意 F.one_hot(tensor, len) 的作用
# tensor: tensor 中的每一个元素都会被编码成 One-Hot 向量
# len:    表示 One-Hot 编码的元素总个数
X = torch.arange(10).reshape((2, 5))
F.one_hot(X.T, 28).shape
```

## 初始化模型参数
需要初始化的参数, 隐藏层(关于隐状态更新)$\mathbf{W}_{xh}, \mathbf{W}_{hh}, \mathbf{b}_h$, 以及输出层(单层输出相关)参数: $\mathbf{W}_{hq}, \mathbf{b}_q$ 等参数

```python
# 初始化模型参数
def get_params(vocab_size, num_hiddens, device):
    num_inputs = vocab_size  # 步长, 特征数
    num_outputs = vocab_size
    def normal(shape):
        return torch.randn(size=shape, device=device) * 0.01
    # 隐藏层参数
    W_xh = normal((num_inputs, num_hiddens))
    W_hh = normal((num_hiddens, num_hiddens))
    b_h = torch.zeros(num_hiddens, device=device) # 一维张量
    # 输出层参数
    W_hq = normal((num_hiddens, num_outputs))
    b_q = torch.zeros(num_outputs, device=device)
    # print('--- get_param ---')
    # print('W_xh shape: ', W_xh.shape)
    # print('W_hh shape: ', W_hh.shape)
    # print('b_h  shape: ', b_h.shape)
    # print('W_hq shape: ', W_hq.shape)
    # print('b_q  shape: ', b_q.shape)
    # print('-----------------')
    # 附加梯度
    params = [W_xh, W_hh, b_h, W_hq, b_q]
    for param in params:
        param.requires_grad_(True)
    return params
```

## 循环神经网络模型

```python
# init_rnn_state: 初始化 RNN 状态, 返回隐状态 n*h
def init_rnn_state(batch_size, num_hiddens, device):
    return (torch.zeros((batch_size, num_hiddens), device=device),)
```

```python
# rnn: 在一个时间步内计算隐状态和输出
def rnn(inputs, state, params):
    W_xh, W_hh, b_h, W_hq, b_q = params
    H, = state
    # X(批量大小, 词表大小)
    outputs = []
    for X in inputs:
        H = torch.tanh(torch.mm(X, W_xh) + torch.mm(H, W_hh) + b_h)
        Y = torch.mm(H, W_hq) + b_q
        outputs.append(Y)
    return torch.cat(outputs, dim=0), (H,) # 注意这就是整个 rnn 网络, 拼接了所有批次的输入
```

```python
# RNNModelScratch: RNN 模型
class RNNModelScratch: #@save
    """循环神经网络模型"""
    def __init__(self, vocab_size, num_hiddens, device, 
                 get_params, init_state, forward_fn):
        self.vocab_size, self.num_hiddens = vocab_size, num_hiddens
        self.params = get_params(vocab_size, num_hiddens, device)
        self.init_state, self.forward_fn = init_state, forward_fn
    def __call__(self, X, state):
        X = F.one_hot(X.T, self.vocab_size).type(torch.float32)
        # print('---call---')
        # print('X shape: ', X.shape)
        # print('----------')
        return self.forward_fn(X, state, self.params)
    def begin_state(self, batch_size, device):
        return self.init_state(batch_size, self.num_hiddens, device)
```

```python
num_hiddens = 512
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
net = RNNModelScratch(len(vocab), num_hiddens, device, get_params,
                     init_rnn_state, rnn)
state = net.begin_state(X.shape[0], device)
Y, new_state = net(X.to(device), state)
Y.shape, len(new_state), new_state[0].shape # 状态列表 
# (时间步数*批量大小, 词表大小) ---> 输出形状
# (批量大小, 隐藏单元数)       ---> 隐状态形状
```

## 预测
定义预测函数用于生成 `prefix` 之后的新字符, 其中 `prefix` 是一个用户提供的包含多个字符的字符串, 遍历 `prefix` 中的开始的字符的时候, 不断将隐状态传递到下一个时间步, 但是不产生任何输出, 也就是预热期 ; 因此在此期间模型会自我更新(更新隐状态), 但是不会进行预测, 预热期结束之后, 隐状态隐状态的值通常比刚开始更加适合预测

```python
def predict_ch8(prefix, num_preds, net, vocab, device):  #@save
    """在prefix后面生成新字符"""
    state = net.begin_state(batch_size=1, device=device)
    outputs = [vocab[prefix[0]]]
    get_input = lambda: torch.tensor([outputs[-1]], device=device).reshape((1, 1))
    for y in prefix[1:]:  # 预热期
        _, state = net(get_input(), state)
        outputs.append(vocab[y])
    for _ in range(num_preds):  # 预测num_preds步
        y, state = net(get_input(), state)
        outputs.append(int(y.argmax(dim=1).reshape(1)))
    return ''.join([vocab.idx_to_token[i] for i in outputs])
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
predict_ch8('time traveller', 10, net, vocab, device)
```

## 梯度裁减
对于长度为 $T$ 的序列, 可以在迭代中计算 $T$ 个时间步上的梯度, 将会在反向传播过程中产生长度为 $O(T)$ 的矩阵乘法链, 并且当 $T$较大的时候, 可能导致导数数值方向不稳定, 从而到时梯度爆炸或者梯度消失, 所以循环神经网络模型需要通过额外的方式来支持稳定训练

一般对于模型参数采用更新步骤, 在向量形式的 $\mathbf{x}$ 中, 或者小批量的数据负梯度 $\mathbf{g}$方向上, 例如使用 $\eta > 0$作为学习率, 在下一次迭代中, 将 $\mathbf{x}$ 更新为 $\mathbf{x} - \eta \mathbf{g}$, 同时如果目标函数 $f$ 表现良好, 也就是函数 $f$ 在常数 $L$ 满足如下条件:
$$
|f(\mathbf{x}) - f(\mathbf{y})| \leq L \|\mathbf{x} - \mathbf{y}\|.
$$
在这一种情况下, 可以安全假设, 如果更新了参数, 那么:
$$
|f(\mathbf{x}) - f(\mathbf{x} - \eta \mathbf{g})| \leq L \eta \|\mathbf{g}\|,
$$
此时意味着不会变换不会超过 $L\eta \|\mathbf{g}\|$, 但是限制了收敛的速度, 另外一个方面, 缓解了梯度消失或者梯度爆炸问题

并且如果梯度过大, 可以通过降低学习率 $\eta$ 的方式来解决这一个问题, 但是如果很少得到大的梯度, 那么此时可以通过将梯度投影回给定半径 $\theta$ 的球来裁减梯度 $\mathbf{g}$, 如下式:
$$
\mathbf{g} \leftarrow \min\left(1, \frac{\theta}{\|\mathbf{g}\|}\right) \mathbf{g}.
$$
通过这样, 梯度的范数永远不会超过 $\theta$, 并且更新之后的梯度完全和 $\mathbf{g}$ 的原始方向对齐, 提供了一个快速修复梯度爆炸的方法(**可以用于缓解梯度爆炸**)

```python
# 梯度裁减实现(用于缓解梯度爆炸)
def grad_clipping(net, theta):
    """梯度裁减"""
    if isinstance(net, nn.Module):
        params = [p for p in net.parameters() if p.requires_grad]
    else:
        params = net.params
    norm = torch.sqrt(sum(torch.sum((p.grad ** 2)) for p in params)) # 表示梯度的L2范数
    if norm > theta:
        for param in params:
            param.grad[:] *= theta/norm # 从而在梯度下降的时候选择合适的梯度
```

## 训练
对于循环神经网络训练的不同之处:
1. 序列数据的不同采样方式(随机采样和顺序分区)将导致隐状态初始化的差异
2. 在更新模型参数之前裁减梯度, 从而避免缓解梯度爆炸(或者防止梯度爆炸传播)
3. 使用困惑度来评价模型, 从而确保不同长度的序列具有可比性

同时使用顺序分区的时候, 指在每一个迭代周期的开始位置初始化隐状态, 由于下一个小批量数据中的第 $i$ 个子序列样本与当前第$i$个子序列样本相邻, 所以当前小批量数据最后一个样本的隐状态用于初始化下一个小批量数据的第一个样本的隐状态(顺序分区下) ; 同时任何一点依状态仅仅依赖于同一个迭代周期中前面的小批量数据, 为了降低复杂度, 在处理任何一个小批量数据之前, 需要是首先分离梯度, 使得隐状态梯度计算限制在一个小批量数据的时间步内

```python
#@save
def train_epoch_ch8(net, train_iter, loss, updater, device, use_random_iter):
    """训练网络一个迭代周期（定义见第8章）"""
    state, timer = None, d2l.Timer()
    metric = d2l.Accumulator(2)  # 训练损失之和,词元数量
    for X, Y in train_iter:
        if state is None or use_random_iter:
            # 在第一次迭代或使用随机抽样时初始化state
            state = net.begin_state(batch_size=X.shape[0], device=device)
        else:
            if isinstance(net, nn.Module) and not isinstance(state, tuple):
                # state对于nn.GRU是个张量
                state.detach_()
            else:
                # state对于nn.LSTM或对于我们从零开始实现的模型是个张量
                # 注意这里每一次都是循环使用相同的参数 W_xh, W_hh
                # 对于新的数据, 原来的数据可以看成标量, 防止重复计算原来的数据
                for s in state:
                    s.detach_()
        y = Y.T.reshape(-1)
        X, y = X.to(device), y.to(device)
        y_hat, state = net(X, state)
        l = loss(y_hat, y.long()).mean()
        if isinstance(updater, torch.optim.Optimizer):
            updater.zero_grad()
            l.backward()
            # 梯度裁减
            grad_clipping(net, 1)
            updater.step()
        else:
            l.backward()
            grad_clipping(net, 1)
            # 因为已经调用了mean函数
            updater(batch_size=1)
        metric.add(l * y.numel(), y.numel())
    return math.exp(metric[0] / metric[1]), metric[1] / timer.stop()
```

```python
#@save
def train_ch8(net, train_iter, vocab, lr, num_epochs, device,
              use_random_iter=False):
    """训练模型（定义见第8章）"""
    loss = nn.CrossEntropyLoss(reduction='none')
    animator = d2l.Animator(xlabel='epoch', ylabel='perplexity',
                            legend=['train'], xlim=[10, num_epochs])
    # 初始化
    if isinstance(net, nn.Module):
        updater = torch.optim.SGD(net.parameters(), lr)
    else:
        updater = lambda batch_size: d2l.sgd(net.params, lr, batch_size)
    predict = lambda prefix: predict_ch8(prefix, 50, net, vocab, device)
    # 训练和预测
    for epoch in range(num_epochs):
        ppl, speed = train_epoch_ch8(
            net, train_iter, loss, updater, device, use_random_iter)
        if (epoch + 1) % 10 == 0:
            print(predict('time traveller'))
            animator.add(epoch + 1, [ppl]) # 注意困惑度计算方法, e^(平均损失(交叉熵))
    print(f'困惑度 {ppl:.1f}, {speed:.1f} 词元/秒 {str(device)}')
    print(predict('time traveller'))
    print(predict('traveller'))
```

```python
num_epochs, lr = 500, 1
train_ch8(net, train_iter, vocab, lr, num_epochs, device)
```

## 定义模型
这一部分用到 `PyTorch API`, 可以使用 `nn.RNN`, 参考: https://docs.pytorch.org/docs/stable/generated/torch.nn.RNN.html

```python
# 定义 RNN 层, 注意对于单个样本, input size 就是 One-Hot 向量长度, len(vocab)
num_hiddens = 256
rnn_layer = nn.RNN(len(vocab), num_hiddens)
```

```python
# 初始化隐状态 shape = (隐藏层数, 批量大小, 隐藏单元数)
state = torch.zeros((1, batch_size, num_hiddens))
state.shape
```

```python
# 测试 X 的 shape, 注意此时输入维度的变换
X = torch.randint(low=0, high=28, size=(batch_size, num_steps))
XH = F.one_hot(X.T, len(vocab))
X.shape, XH.shape
```

```python
# 演示 rnn_layer 作用, 需要输入一个 X, 以及一个隐藏层状态
# num_steps 时间步, batch_size 批量, len(vocab): input size
X = torch.rand(size=(num_steps, batch_size, len(vocab)))
# Y 表示每一个时间步的隐状态, state_new 表示最后一个时间步的隐状态
Y, state_new = rnn_layer(X, state) # 相当于 35 单元, 注意这里的 Y 不涉及输出层的计算, 
Y.shape, state_new.shape
```

```python
# 定义 RNNModel 类, rnn_layer 只包含隐藏层的循环层, 可以啊记入一个单独的输出层
#@save
class RNNModel(nn.Module):
    """循环神经网络模型"""
    def __init__(self, rnn_layer, vocab_size, **kwags):
        super(RNNModel, self).__init__(**kwags)
        self.rnn = rnn_layer
        self.vocab_size = vocab_size
        self.num_hiddens = self.rnn.hidden_size
        # 判断 RNN 是否是双向的
        if not self.rnn.bidirectional:
            self.num_directions = 1
            self.linear = nn.Linear(self.num_hiddens, self.vocab_size)
        else:
            self.num_directions = 2
            self.linear = nn.Linear(self.num_hiddens*2, self.vocab_size)
    def forward(self, inputs, state):
        X = F.one_hot(inputs.T.long(), self.vocab_size) # 注意输入变换
        X = X.to(torch.float32)
        Y, state = self.rnn(X, state)
        # 注意这里加上一个线性层表示最后需要使用 H_t 得到对应的输出
        # 首先把 Y 形状变成 (时间步数*批量大小, 隐藏单元数)
        output = self.linear(Y.reshape((-1, Y.shape[-1])))
        return output, state
    def begin_state(self, device, batch_size=1):
        if not isinstance(self.rnn, nn.LSTM):
            # nn.GRU以张量作为隐状态
            return  torch.zeros((self.num_directions * self.rnn.num_layers,
                                 batch_size, self.num_hiddens),
                                device=device)
        else:
            # nn.LSTM以元组作为隐状态
            return (torch.zeros((
                self.num_directions * self.rnn.num_layers,
                batch_size, self.num_hiddens), device=device),
                    torch.zeros((
                        self.num_directions * self.rnn.num_layers,
                        batch_size, self.num_hiddens), device=device))
```

## 训练以及预测

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
net = RNNModel(rnn_layer, vocab_size=len(vocab))
net = net.to(device)
predict_ch8('time traveller', 10, net, vocab, device)
```

```python
num_epochs = 500
lr = 1
train_ch8(net, train_iter, vocab, lr, num_epochs, device)
```
