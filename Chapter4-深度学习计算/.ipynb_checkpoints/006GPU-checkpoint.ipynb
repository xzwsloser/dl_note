
上述内容中的显示相关信息如下:
**整体头部信息:**
- `NVIDIA-SMI`: 工具自带的版本号
- `Driver Version`: `GPU`驱动版本
- `CUDA Version`: 表示 `CUDA` 版本

**GPU设备基本状态表格:**
- `GPU`: 设备编号(比如两块 `GPU` 编号为 `0、1`)
- `Name`: `GPU`具体型号
- `Persistence-M`: 持久模式状态(是否加载 `GPU` 驱动到内存中)
- `Bus-Id`: `GPU` 在 `PCle` 总线上面的位置标识
- `Disp.A`: 标识是否用于显示输出
- `Volatile Uncorr.ECC`: ...
- `Fan`: 风扇状态
- `Temp`: `GPU`核心温度
- `Perf`: 性能状态(`P0`标识最高状态, 数字越大, 性能越低下)
- `Pwr`: 工具使用情况
- `Memory-Usage`: 显存使用状态
- `GPU-Utils`: `GPU` 利用率
- `Compute M`: 计算模式
- `MIG M`: 把一块物理 `GPU` 划分为多个实例

**进程信息表格:**
展示了各种进程信息, 其中 `GID`感觉比较有意思

## 计算设备
可以指定计算设备, 默认情况下张量在内存中创建, 并且可以使用 `CPU` 计算张量, 在 `PyTorch` 中, 可以使用 `torch.device('cpu'), torch.device('cuda')`来表示 `CPU` 和 `GPU`, 同时如果多张 `GPU`, 可以使用 `torch.device(f'cuda{i}')`表示第 `i` 块 `GPU`

```python
import torch
import torch.nn as nn
torch.device('cpu'), torch.device('cuda'), torch.device('cuda:1')
```

```python
# 可以查询可用的 GPU 数量
torch.cuda.device_count()
```

```python
def try_gpu(i=0):  #@save
    """如果存在，则返回gpu(i)，否则返回cpu()"""
    if torch.cuda.device_count() >= i + 1:
        return torch.device(f'cuda:{i}')
    return torch.device('cpu')

def try_all_gpus():  #@save
    """返回所有可用的GPU，如果没有GPU，则返回[cpu(),]"""
    devices = [torch.device(f'cuda:{i}')
             for i in range(torch.cuda.device_count())]
    return devices if devices else [torch.device('cpu')]

try_gpu(), try_gpu(10), try_all_gpus()
```

## 张量与GPU
可以查询张量所在的设备, 默认情况下, 张量在 `CPU` 上创建

```python
x = torch.tensor([1, 2, 3])
x.device
```

```python
# 1. 存储张量在 GPU 上
X = torch.ones(2, 3, device=try_gpu())
X
```

```python
Y = torch.rand(2, 3, device=try_gpu(1))
Y
```

### 复制
比如需要计算 X + Y, 那么此时会把一个张量复制到另外一个 `GPU` 上

```python
Z = X.cuda(1) # 表示把 X 复制到 cuda1 的 Z 上
print(X)
print(Y)
```

```python
# 判断张量
Z.cuda(1) is Z
```

## 神经网络与 GPU

```python
net = nn.Sequential(nn.Linear(3, 1))
net = net.to(device=try_gpu())
```

```python
net(X)
```

```python
# 注意确认模型参数存储在同一个 GPU 上
net[0].weight.data.device
```
