# softmax回归
> 线性回归问题用于解决回归问题, softmax回归问题用于解决分类问题
## 分类问题
首先还是考虑单个样本, 假设单个样本具有 $4$ 个特征, 也就是 $x = {[x_1, x_2, x_3, x_4]}^T$, 最后需要把样本分为三类, 假设为 $o_1, o_2, o_3$, 此时可以使用独热编码来表示这三个分类, 也就是 $o_1 = {[1, 0, 0]}^T$, $o_2  = {[0, 1, 0]}^T$, $o_3 = {[0, 0, 1]}^T$, 所以标签 $y \in \{o_1, o_2, o_3\}$
## 网络架构
此时可以使用全连接层来表示这一个网络结构, 网络结构如下:

那么此时输入层到输出层的映射关系可以表示为:
$$
\begin{align*}
o_1 &= x_1w_{11} + x_2w_{12} + x_3w_{13} + x_4w_{14} + b_1, \\
o_2 &= x_1w_{21} + x_2w_{22} + x_3w_{23} + x_4w_{24} + b_2, \\
o_3 &= x_1w_{31} + x_2w_{32} + x_3w_{33} + x_4w_{34} + b_3.
\end{align*}
$$
向量形式可以表示为:
$$
\mathbf{o} = \mathbf{W}\mathbf{x} + \mathbf{b}
$$
假设特征数为 $k$, 分类数为 $m$, 那么此时 $\mathbf{o}$ 为 $m \times 1$, $\mathbf{W}$ 为 $m \times k$, 此时 $\mathbf{x}$ 为 $k \times 1$, $\mathbf{b}$ 为 $m \times 1$
## softmax 运算
对于分类问题, 最后选取的分类一定是出现概率最大的一个分类, 但是无法直接根据没有泛化的输出 $\mathbf{o}$ 来决定预测的分类, 所以可以使用 softmax 运算对于输出进行泛化(归一化等操作), 定义 softmax 运算如下:
$$
y_{pred} = \frac{exp(o_j)}{\sum_k exp(o_k)}
$$
那么就有:
$$
\mathbf{y}_{pred} = softmax(\mathbf{o})
$$
那么此时把输出泛化为 $\mathbf{y}_{pred}$, 此时就是一个正确的概率分布, 同时在, 预测过程中使用如下方式就可以找到正确的类别:
$$
\underset{j}{\text{argmax}} \hat{y}_j = \underset{j}{\text{argmax}} o_j
$$
虽然 softmax 是一个非线性函数, 但是 softmax 回归的输出需要由输入特征的仿射变换决定, 所以依然仍为 softmax 回归是一个线性模型
## 小批量样本矢量化
上面都是讨论了单个样本的 softmax 回归公式, 但是在实际问题中一般每一次优化都需要采用一批样本, 那么假设对于一个批量的样本 $\mathbf{X}$, 特征维度为 $d$, 批量大小为 $n$, 假设此时输入中含有 $q$ 个类别, 那么此时 $\mathbf{X} \in \mathbb{R}^{n \times d}$, 权重 $\mathbf{W} \in \mathbb{R}^{d \times q}$ , 偏置 $\mathbf{b} \in \mathbb{R}^{1 \times q}$, 此时 softmax 回归的矢量计算公式如下:
$$
\mathbf{O} = \mathbf{X}\mathbf{W} + \mathbf{b}, \\\
\mathbf{\hat{Y}} = softmax(\mathbf{O})
$$
在 `PyTorch` 中进行运算的时候此时 $\mathbf{b}$ 会执行广播运算, 从而拓展到 $n \times q$
## 损失函数
在 softmax 回归问题中使用交叉熵函数作为损失函数
### 对数似然
注意最大似然估计的意义是求解当输入为 $\mathbf{X}$ 的时候, 输出为对应的 $\mathbf{Y}$ 的最大概率, 其中 $\{\mathbf{X}, \mathbf{Y}\}$ 总共涉及 $n$ 个样本, 那么此时需要最大化的概率如下:
$$
P(\boldsymbol{Y} \mid \boldsymbol{X}) = \prod_{i=1}^{n} P(\boldsymbol{y}^{(i)} \mid \boldsymbol{x}^{(i)}).
$$
此时需要最大化 $P{\mathbf{Y} \mid \mathbf{X}}$, 也就是最小化负对数似然:
$$
-\log P(\boldsymbol{Y} \mid \boldsymbol{X}) = \sum_{i=1}^{n} - \log P(\boldsymbol{y}^{(i)} \mid \boldsymbol{x}^{(i)}) = \sum_{i=1}^{n} l(\boldsymbol{y}^{(i)}, \hat{\boldsymbol{y}}^{(i)}),
$$
对于任何标签 $\mathbf{y}$ 以及预测 $\hat{\mathbf{Y}}$, 损失函数使用交叉熵损失:
$$
l(\boldsymbol{y}, \hat{\boldsymbol{y}}) = - \sum_{j=1}^{q} y_j \log \hat{y}_j.
$$
### softmax及其导数
下面考虑损失函数和各个参数之间梯度的关系, 对于损失函数进行恒等变形可以得到:
$$
\begin{align*}
l(\boldsymbol{y}, \hat{\boldsymbol{y}}) &= -\sum_{j=1}^{q} y_j \log \frac{\exp(o_j)}{\sum_{k=1}^{q} \exp(o_k)} \\
&= \sum_{j=1}^{q} y_j \log \sum_{k=1}^{q} \exp(o_k) - \sum_{j=1}^{q} y_j o_j \\
&= \log \sum_{k=1}^{q} \exp(o_k) - \sum_{j=1}^{q} y_j o_j.
\end{align*}
$$
那么对于 $o_j$ 求解导数可以的得到:
$$
\partial_{o_j} l(\boldsymbol{y}, \hat{\boldsymbol{y}}) = \frac{\exp(o_j)}{\sum_{k=1}^{q} \exp(o_k)} - y_j = \text{softmax}(\boldsymbol{o})_j - y_j.
$$
所以此时就可以就可以使用反向传播求解出损失函数对于各个参数的梯度从而进行梯度下降操作了
### 交叉熵损失
在信息论中, 熵其实是表示不确定度, 熵越小表示不确定度越小, 预测越准确 ; 从信息论的角度去理解交叉熵可以这样理解 $H(P)$ 表示事件 $P$ 的不确定度, 那么交叉熵$H(P , Q)$ 表示主观概率为 $Q$ 的观察者在看到根据概率 $P$ 生成数据时的不确定度(其实就是 KL 散度), 如果 $P == Q$, 此时交叉熵最小, 也就是 $H(P, P) = H(P)$, 其实由香农不等式可以得到: $H(P,Q) \geq H(P)$

```python
# 1. 读取数据集
# ToTensor 函数把图像所有像素的数组映射到 0 - 1
trans = transforms.ToTensor()
mnist_train = torchvision.datasets.FashionMNIST(
    root='./data', train=True, transform=trans, download=True
)

mnist_test = torchvision.datasets.FashionMNIST(
    root='./data', train=False, transform=trans, download=True
)
```

```python
# Fashion-MNIST 包括 10 个类别
def get_fashion_mnist_labels(labels):  #@save
    """返回Fashion-MNIST数据集的文本标签"""
    text_labels = ['t-shirt', 'trouser', 'pullover', 'dress', 'coat',
                   'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot']
    return [text_labels[int(i)] for i in labels]
```

```python
# 显示图像
def show_images(imgs, num_rows, num_cols, titles=None, scale=1.5):
    """绘制图像列表"""
    figsize = (num_cols*scale, num_rows*scale)
    _, axes = plt.subplots(num_rows, num_cols, figsize=figsize) # 创建子图 axes 对象
    axes = axes.flatten() # 展开为一维数组
    for i, (ax, img) in enumerate(zip(axes, imgs)):  # zip 可以生成一个迭代器
        if torch.is_tensor(img):
            ax.imshow(img.numpy())
        else:
            ax.imshow(img)
        ax.axes.get_xaxis().set_visible(False)
        ax.axes.get_yaxis().set_visible(False)
        if titles:
            ax.set_title(titles[i])
    return axes
```

# softmax回归实现(造轮子)

```python
# 0. 定义数据集加载操作
def load_data_fashion_mnist(batch_size):
    trans = transforms.Compose([transforms.ToTensor()])
    mnist_train = torchvision.datasets.FashionMNIST(
        root='./data',train=True,transform=trans,download=True
    )
    mnist_test = torchvision.datasets.FashionMNIST(
        root='./data',train=False,transform=trans,download=True
    )
    return data.DataLoader(mnist_train, batch_size, shuffle=True), data.DataLoader(mnist_train, batch_size, shuffle=False)
```

```python
# 1. 准备工作
batch_size=256
train_iter, test_iter = load_data_fashion_mnist(batch_size)
```

```python
# 2. 初始化模型参数
# 考虑输入为 28*28 个特征, 输出为 10 个分类
num_inputs = 784
num_outputs = 10
W = torch.normal(0, 0.01, (num_inputs, num_outputs),requires_grad=True)
b = torch.zeros(num_outputs,requires_grad=True)
```

```python
# 2. 定义 softmax 操作
def softmax(X):
    X_exp = torch.exp(X)
    parition = X_exp.sum(1, keepdims=True) # 保留维度
    return X_exp / parition
X = torch.normal(0, 1, (2, 5))
X_prob = softmax(X)
X_prob, X_prob.sum(1)
```

```python
# 3. 定义模型
def net(X):
    # 把图像展开为像素点, 之后进行线性运算
    return softmax(torch.matmul(X.reshape((-1, W.shape[0])), W) + b)  
```

```python
# 4. 定义损失函数, 注意此时 y 使用 one-hot 编码, 所以此时损失就是 -log P
def cross_entropy(y_pred, y):
    return -torch.log(y_pred[range(len(y_pred)), y])
y = torch.tensor([0, 2]) # 真实数据
y_pred = torch.tensor([[0.1, 0.3, 0.6],[0.3, 0.2, 0.5]])
cross_entropy(y_pred, y)
```

```python
# 5. 分类精度, 也就是准确率
def accuracy(y_hat, y):  #@save
    """计算预测正确的数量"""
    if len(y_hat.shape) > 1 and y_hat.shape[1] > 1:
        y_hat = y_hat.argmax(axis=1)
    cmp = y_hat.type(y.dtype) == y
    return float(cmp.type(y.dtype).sum())
# 累加值
class Accumulator:  #@save
    """在n个变量上累加"""
    def __init__(self, n):
        self.data = [0.0] * n

    def add(self, *args):
        self.data = [a + float(b) for a, b in zip(self.data, args)]

    def reset(self):
        self.data = [0.0] * len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]
# 评估准确率
def evaluate_accuracy(net, data_iter):  #@save
    """计算在指定数据集上模型的精度"""
    metric = Accumulator(2)  # 正确预测数、预测总数
    for X, y in data_iter:
        metric.add(accuracy(net(X), y), y.numel()) # 后者表示元素总数
    return metric[0] / metric[1]
# 评估初始权重
evaluate_accuracy(net, test_iter)
```

```python
# 6. 开始训练
def train_epoch_ch3(net, train_iter, loss, updater):  #@save
    """训练模型一个迭代周期（定义见第3章）"""
    # 将模型设置为训练模式
    if isinstance(net, torch.nn.Module):
        net.train()
    # 训练损失总和、训练准确度总和、样本数
    metric = Accumulator(3)
    for X, y in train_iter:
        # 计算梯度并更新参数
        y_hat = net(X)
        l = loss(y_hat, y)
        if isinstance(updater, torch.optim.Optimizer):
            # 使用PyTorch内置的优化器和损失函数
            updater.zero_grad()
            l.mean().backward()
            updater.step()
        else:
            # 使用定制的优化器和损失函数
            l.sum().backward()
            updater(X.shape[0])
        metric.add(float(l.sum()), accuracy(y_hat, y), y.numel())
    # 返回训练损失和训练精度
    return metric[0] / metric[2], metric[1] / metric[2]
```

```python
# 一个关于绘制的函数
class Animator:  #@save
    """在动画中绘制数据"""
    def __init__(self, xlabel=None, ylabel=None, legend=None, xlim=None,
                 ylim=None, xscale='linear', yscale='linear',
                 fmts=('-', 'm--', 'g-.', 'r:'), nrows=1, ncols=1,
                 figsize=(3.5, 2.5)):
        # 增量地绘制多条线
        if legend is None:
            legend = []
        d2l.use_svg_display()
        self.fig, self.axes = d2l.plt.subplots(nrows, ncols, figsize=figsize)
        if nrows * ncols == 1:
            self.axes = [self.axes, ]
        # 使用lambda函数捕获参数
        self.config_axes = lambda: d2l.set_axes(
            self.axes[0], xlabel, ylabel, xlim, ylim, xscale, yscale, legend)
        self.X, self.Y, self.fmts = None, None, fmts

    def add(self, x, y):
        # 向图表中添加多个数据点
        if not hasattr(y, "__len__"):
            y = [y]
        n = len(y)
        if not hasattr(x, "__len__"):
            x = [x] * n
        if not self.X:
            self.X = [[] for _ in range(n)]
        if not self.Y:
            self.Y = [[] for _ in range(n)]
        for i, (a, b) in enumerate(zip(x, y)):
            if a is not None and b is not None:
                self.X[i].append(a)
                self.Y[i].append(b)
        self.axes[0].cla()
        for x, y, fmt in zip(self.X, self.Y, self.fmts):
            self.axes[0].plot(x, y, fmt)
        self.config_axes()
        display.display(self.fig)
        display.clear_output(wait=True)
```

```python
# 训练模型
def train_ch3(net, train_iter, test_iter, loss, num_epochs, updater):  #@save
    """训练模型（定义见第3章）"""
    animator = Animator(xlabel='epoch', xlim=[1, num_epochs], ylim=[0.3, 0.9],
                        legend=['train loss', 'train acc', 'test acc'])
    for epoch in range(num_epochs):
        train_metrics = train_epoch_ch3(net, train_iter, loss, updater)
        test_acc = evaluate_accuracy(net, test_iter)
        animator.add(epoch + 1, train_metrics + (test_acc,))
    train_loss, train_acc = train_metrics
    assert train_loss < 0.5, train_loss
    assert train_acc <= 1 and train_acc > 0.7, train_acc
    assert test_acc <= 1 and test_acc > 0.7, test_acc
```

```python
# 定义优化器函数
lr = 0.1

def updater(batch_size):
    return d2l.sgd([W, b], lr, batch_size)

# 开始训练
num_epochs = 10
train_ch3(net, train_iter, test_iter, cross_entropy, num_epochs, updater)
```

```python
# 预测
def predict_ch3(net, test_iter, n=6):  #@save
    """预测标签（定义见第3章）"""
    for X, y in test_iter:
        break
    trues = d2l.get_fashion_mnist_labels(y)
    preds = d2l.get_fashion_mnist_labels(net(X).argmax(axis=1))
    titles = [true +'\n' + pred for true, pred in zip(trues, preds)]
    d2l.show_images(
        X[0:n].reshape((n, 28, 28)), 1, n, titles=titles[0:n])

predict_ch3(net, test_iter)
```
