# 线性回归
在考虑一种机器学习或者深度学习方法的时候需要关注几个部分:
1. 模型结构(或者网络结构)
2. 损失函数
3. 优化器
## 线性模型
对于单个样本 $\mathbf{x}$, 假设 $\mathbf{x}$ 有  $d$ 个特征, 那么 $\mathbf{x} = {[x_1, x_2, ..., x_d]}$ , 假设此时的预测值为 $\mathbf{y}_{pred}$, 其中 $\mathbf{y}_{pred}$ 有 $n$ 个特征, 也就是 $\mathbf{y}_{pred} = {[y_1, y_2, ..., y_n]}^T$, 那么需要需要得到一个权重参数 $\mathbf{w}$, 大小为 $n \times d$, 另外还有一个偏置参数 $\mathbf {b}$, 对于单个样本, 线性模型为:
$$
\mathbf{y}_{pred} = \mathbf{w}\mathbf{x} + \mathbf{b}
$$
其中对于 $w_{ij}$ 表示 特征 $x_j$ 对于 $y_i$ 的权重影响

如果考虑多个样本, 假设有 $n$ 个样本, 那么使用矩阵 $\mathbf{X}$ 表示样本矩阵, 大小为 $n \times d$, 那么此时得到的预测值为 $\mathbf{y}_{pred}$ 大小为 $n \times m$(假设有 $m$ 个特征), 那么此时的权重参数 $\mathbf{w}$ 的大小为 $d \times m$, 偏置参数为 $m$, 那么此时线性模型为:

$$
\mathbf{y}_{pred} = \mathbf{X}\mathbf{w} + \mathbf{b}
$$
## 损失函数
损失函数用于描述预测值与真实值之间的误差, 是一个确定拟合程度的度量, 可以量化目标的实际值和预测值之间的差距, 比如平方误差函数就是一种常用的损失函数, 平方误差函数定义如下:
对于样本 $i$ , 预测值为 ${y_{pred}}^{(i)}$, 实际值为 $y^{(i)}$, 平方误差定义为
$$
l^{(i)}(\mathbf{w}, b) = \frac{1}{2} ({y_{pred}}^{(i)} - y^{(i)})^2
$$

对于整个训练集上的 $n$ 个样本, 损失均值为:
$$
L(\mathbf{w}, b) = \frac{1}{n} \sum_{i=1}^{n} l^{(i)}(\mathbf{w}, b) = \frac{1}{n} \sum_{i=1}^{n} \frac{1}{2} \bigl( \mathbf{w}^{\top} \mathbf{x}^{(i)} + b - y^{(i)} \bigr)^{2}
$$
训练模型的时候, 就是希望找到一组参数 $(\mathbf{w}^*, b^*)$ 使得上述的损失函数值最小

## 随机梯度下降
> 随机梯度下降是一种优化器函数

注意到如果让上述的 $L(\mathbf{w}, b)$ 最小, 只需要让对应位置的导数为 $0$ 了, 这一个解表示整个解析解, 但是有时候一些问题不存在解析解, 所以需要使用优化器求解最优解, 随机梯度下降就是一种优化器方法

如果直接使用梯度下降方法, 需要每一次都遍历整个训练集, 为了提高效率实际可以使用随机梯度下降, 也就是每一次知考虑一批样本, 样本数量为 $\mathbf{B}$ , 对于这一批数据使用梯度下降称为小批量梯度下降(也叫做随机梯度下降), 更新过程如下:
$$
(\mathbf{w}, b) \leftarrow (\mathbf{w}, b) - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} \partial_{(\mathbf{w}, b)} l^{(i)}(\mathbf{w}, b)
$$

算法的步骤:
1. 初始化模型参数的值(`PyTorch` 中会自动初始化模型参数)
2. 从数据集中随机抽取小批量样本并且在负梯度方向更新参数并且不断迭代, 更新过程如下:
$$
\mathbf{w} \leftarrow \mathbf{w} - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} \partial_{\mathbf{w}} l^{(i)}(\mathbf{w}, b) = \mathbf{w} - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} \mathbf{x}^{(i)} \bigl( \mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)} \bigr)  
$$

$$
b \leftarrow b - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} \partial_{b} l^{(i)}(\mathbf{w}, b) = b - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} \bigl( \mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)} \bigr)
$$

其中 $\eta$ 表示学习率, 是我们自己规定的一个参数, 不是通过学习模型训练得到的,这种参数称为**超参数**, 调整超参数的过程称为**调参**

模型在完全陌生的数据集上的效果称为**泛化能力**

## 正态分布与平方损失
> 参考: https://zh.d2l.ai/chapter_linear-networks/linear-regression.html

可以证明: 在高斯噪声的假设下, 最小化均方误差等价于对线性模型的极大似然估计(没有高斯噪声的条件无法估算概率)

## 从线性回归到深度网络
线性回归可以看成一个简单的单层神经网络:

对于线性回归, 每一个输入都和每一个输出相联, 在这一种变化中输出层称为**全连接层**
### 神经元
神经网络中的每一个节点相当于一个神经元, 一个神经元接受来自其他神经元的信息 $x_i$ , 并且通过突触权重 $w_i$ 进行加权, 输出就是 $y = \sum_{i}x_i w_i + b$

# 线性回归实现(造轮子)

```python
# 生成数据集
def generate_data(w, b, num_examples):
    """
    w: 实际权重
    b: 实际偏置
    num_examples: 样本数量
    """
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1)) # -1 表示这一个维度大小自动推断
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = generate_data(true_w, true_b, 1000)
print('features:', features[0], '\nlabel', labels[0])
```

```python
#  绘制生成的数据
plt.scatter(features[:,1].detach().numpy(), labels.detach().numpy())
plt.show()
```

```python
# 读取数据集, 注意迭代器写法
def data_iter(batch_size, features, labels):
    """返回数据集"""
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0 , num_examples, batch_size): # 步长为 batch_size
        batch_indices = torch.tensor(indices[i:min(i+batch_size,num_examples)])
        yield features[batch_indices], labels[batch_indices]

batch_size = 10
for X,y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

```python
# 初始化模型参数
w = torch.normal(0, 0.01, size=(2, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# 定义模型
def linear(X, w, b):
    """线性回归模型"""
    return torch.matmul(X, w) + b

# 定义损失函数
def squared_loss(y_pred, y):
    """损失函数"""
    return (y_pred - y)**2/2

# 定义优化算法
def sgd(params, lr, batch_size):
    """随机梯度下降"""
    with torch.no_grad():
        for param in params:
            param -= lr*param.grad/batch_size
            param.grad.zero_()
```

```python
# 训练模型
lr = 0.03
num_epoch = 3
net = linear
loss = squared_loss

for epoch in range(num_epoch):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y)
        l.sum().backward() # 反向传播
        sgd([w, b], lr, batch_size) # 梯度下降
    with torch.no_grad():
        # 此时不需要计算梯度, 已经遍历整个训练集
        train_l = loss(net(features, w, b), labels) # 整个训练集上的损失
        print(f'epoch {epoch + 1}, loss{float(train_l.mean()):f}')
        print(f'w = {w} \n b = {b}')
```

# 线性回归简单实现

```python
# 1. 构建数据集
class LinearDataset(Dataset):
    def __init__(self, w, b, num_examples):
        self.x = torch.normal(0, 1, (num_examples, len(w)))
        self.y = torch.matmul(self.x, w) + b
        self.y += torch.normal(0, 0.01, size=self.y.shape)  # 添加噪声
        self.y = self.y.reshape((-1, 1))
    def __len__(self):
        return len(self.y)
    def __getitem__(self, idx):
        x = self.x[idx]
        y = self.y[idx]
        return x, y
true_w = torch.tensor([2.0, -3.4])
true_b = 1.0
dataset = LinearDataset(true_w, true_b, 2000)
dataloader = DataLoader(dataset, batch_size=50, shuffle=True)
```

```python
# 2. 构建模型
net = nn.Sequential(
    nn.Linear(2, 1)
)

# 3. 损失函数
loss_func = nn.MSELoss()
# 4. 优化器
sgd = optim.SGD(net.parameters(), lr=0.03)
```

```python
# 5. 模型训练
loss_sum = 0
for epoch in range(10):
    for features, labels in dataloader:
        y_pred = net(features)
        loss = loss_func(y_pred, labels)
        loss_sum += loss.item()
        sgd.zero_grad() # 清除梯度
        loss.backward() # 反向传播
        sgd.step()   # 更新参数    
    print(f'epoch: {epoch + 1}, loss: {loss_sum/1000}')
    loss_sum = 0
    print(f'w = {net[0].weight.data}, b = {net[0].bias.data}')
```
