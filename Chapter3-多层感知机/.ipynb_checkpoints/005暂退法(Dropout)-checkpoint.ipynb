# 暂退法(Dropout)
## 重新审视过拟合
面对过多的特征但是样本不足的情况下, 线性模型可能出现过拟合的现象, 如果给出更多样本而不是特征, 通常线性模型不会出现过拟合现象, **但是线性模型的缺陷在于没有考虑不同特征之间的相互作用关系**, **之前提出的权重衰退主要就是用于解决线性模型的过拟合问题**

**泛化性和灵活性的度量**: **偏差-方差衡量**, 其中偏差表示泛化能力, 方差表示灵活性, 比如对于线性模型, 可能有比较大的偏差(泛化能力低), 但是可能方差比较小(在随机数据样本上可以得到相似的结果)

对于深度神经网络模型, 即使拥有比特征多得多的样本, 深度神经网络也可能出现过拟合现象, 这里提出的暂退法可以用于解决深度神经网络中出现的过拟合问题
## 扰动的稳健性
对于一个好的预测模型,根据经典泛化理论, 为了缩小训练和测试性能之间的差距, 需要使用简单的模型

同时一个简单模型包括两个方面的意思:
1. 首先简单模型需要以比较小的维度的形式展示, 比如之前提到的权重衰退通过引入 $L_2$ 范数惩罚的方式来尽量降低权重
2. 另外一个维度就是简单模型需要具有平滑性, 也就是函数不应该对于其输入的微小变换敏感

**暂退法用于保证第二个维度上的简单性**: 暂退法在钱向, 传播的过程中, 计算每一层的同时注入噪声, 从而在输入-输出映射上增加平滑性 ; 这一种方法之所以叫做暂退法, **是因为在表面上看是在训练过程中丢弃了一些神经元, 在整个训练的每一次迭代中, 标准暂退法包括在计算下一层之前将当前层中的一些节点置零(相当于丢弃当前节点的输出)**

注入噪声的方式使用无偏向的方式注入噪声, 也就是固定住其他层, 每一层的期望等于没有噪声时的值

具体的实现原来就是住于均值为 0 的噪声 $\epsilon \sim \mathcal{N}(0, \sigma^2)$, 那么此时 $E[x] = E[x + \epsilon]$

在标准暂退法正则化的过程中, 通过按保留(未丢弃)的节点的分数进行规范化来消除每一层的偏差, 也就是每一个中间活性值 $h$ 以暂退概率 $p$ 由随机变量 $h'$ 替换
$$
h' = 
\begin{cases} 
0 & \text{概率为 } p \\
\dfrac{h}{1 - p} & \text{其他情况}
\end{cases}
$$
那么此时依然可以保证: $E[h'] = E[h]$

## 暂退法实现(造轮子版)

```python
# dropout_layer: 以 dropout 概率丢弃张量输入 X 中的元素, 剩余部分除以 1 - dropout
# mask 生成不够严谨
def dropout_layer(X, dropout):
    assert 0 <= dropout <= 1
    if dropout == 1:
        return torch.zeros_like(X)
    if dropout == 0:
        return X
    mask = (torch.rand(X.shape) > dropout).float() # 均匀采样
    return mask*X / (1.0 - dropout)
```

```python
# 假设整个模型的输入为 n*d(d为特征数)
X = torch.tensor([[1, 2],
                  [3, 4]])
dropout_layer(X, 0.5), X
```

```python
# 定义模型参数
# 网络结构:
# 输入层:  n*num_inputs
# 隐藏层1: n*num_hiddens1
# 隐藏层2: n*num_hiddens2
# 输出层:  n*num_outputs
num_inputs, num_outputs, num_hiddens1, num_hiddens2 = 784, 10, 256, 256
```

```python
# 定义模型
# 定义模型的衰退率
dropout1, dropout2 = 0.2, 0.5

class Net(nn.Module):
    """模型训练时使用暂退法, 模型测试时不使用"""
    def __init__(self, num_inputs, num_outputs, num_hiddens1, num_hiddens2,
                is_training=True):
        super(Net, self).__init__()
        self.num_inputs = num_inputs
        self.traning = is_training
        self.lin1 = nn.Linear(num_inputs, num_hiddens1)
        self.lin2 = nn.Linear(num_hiddens1, num_hiddens2)
        self.lin3 = nn.Linear(num_hiddens2, num_outputs)
        self.relu = nn.ReLU()
    def forward(self, X):
        H1 = self.relu(self.lin1(X.reshape((-1, self.num_inputs)))) # 特征数
        if self.training:
            H1 = dropout_layer(H1, dropout1) # 首先 relu, 之后 dropout
        H2 = self.relu(self.lin2(H1))
        if self.training:
            H2 = dropout_layer(H2, dropout2)
        out = self.lin3(H2)
        return out
    # def forward(self, X):
    #     H1 = dropout_layer(self.lin1(X.reshape((-1, self.num_inputs))), dropout1)
    #     if self.training:
    #         H1 = self.relu(H1)
    #     H2 = dropout_layer(self.lin2(H1), dropout2)
    #     if self.training:
    #         H2 = self.relu(H2)
    #     return self.lin3(H2)
net = Net(num_inputs, num_outputs, num_hiddens1, num_hiddens2)
```

```python
# 测试和训练
num_epochs, lr, batch_size=10, 0.5, 256
loss = nn.CrossEntropyLoss(reduce='none')
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
trainer = torch.optim.SGD(net.parameters(), lr=lr)
mdl.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```

## 简洁实现

```python
net = nn.Sequential(
    nn.Flatten(),
    nn.Linear(num_inputs, num_hiddens1),
    nn.ReLU(),
    nn.Dropout(dropout1), # dropout 层
    nn.Linear(num_hiddens1, num_hiddens2),
    nn.ReLU(),
    nn.Dropout(dropout2),
    nn.Linear(num_hiddens2, num_outputs)
)

net1 = nn.Sequential(
    nn.Flatten(),
    nn.Linear(num_inputs, num_hiddens1),
    nn.ReLU(),
    nn.Linear(num_hiddens1, num_hiddens2),
    nn.ReLU(),
    nn.Linear(num_hiddens2, num_outputs)
)
```

```python
# 模型训练
trainer = torch.optim.SGD(net.parameters(), lr=lr)
mdl.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```
